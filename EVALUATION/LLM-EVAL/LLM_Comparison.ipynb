# Read the datafile

```python
import pandas as pd
#df = pd.read_csv("OUTPUT-OLLAMA/MODEL-METRICS-COMPARISON.csv")
df = pd.read_csv("OUTPUT NEW/MODEL-METRICS-COMPARISON BKP.csv")
df.reset_index(drop=True)
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df
```

# Means compared

```python
mean_cols = [col_name for col_name in df.columns if col_name.endswith("-mean_value") or col_name == "metric"]

#df_mean  = df[["metric", "gemma-mean_value", "llama-mean_value","mistral-mean_value"]].copy()
df_mean  = df[mean_cols].copy()
```

```python
import numpy as np

FUNC_MAP = {
    "entity_count_correctly_detected_pct"       : np.nanmax,
    "entity_count_detected_pct"					: np.nanmax,
    "entity_count_undetected_pct"				: np.nanmin,
    "entity_count_wrongly_detected_pct"			: np.nanmin,
    "rel_count_correctly_detected_skip_type_pct": np.nanmax,
    "rel_count_detected_pct"					: np.nanmax,
    "rel_count_entities_correctly_detected_pct" : np.nanmax,
    "rel_count_undetected_pct"					: np.nanmin,
    "rel_count_undetected_skip_type_pct"		: np.nanmin,
    "rel_count_wrongly_detected_pct"            : np.nanmin,
    "rel_count_wrongly_detected_skip_type_pct"  : np.nanmin,
    "time_taken_per_word"                       : np.nanmin
}

from numbers import Number
def highlight(s, function_map={}, color=''):
    func = function_map[s[0]]
    max = func(
                [val for val in s.values if isinstance(val, Number)]
                 )
    return np.where(s==max, color, '')

df_mean.style.apply(highlight, axis=1, function_map=FUNC_MAP, color='background-color:lightgreen;')
```

# Medians compared

```python
median_cols = [col_name for col_name in df.columns if col_name.endswith("-median_value") or col_name == "metric"]
#df_median  = df[["metric", "gemma-median_value", "llama-median_value","mistral-median_value"]].copy()
df_median  = df[median_cols].copy()

df_median.style.apply(highlight, axis=1, function_map=FUNC_MAP, color='background-color:lightgreen;')
```

# Standard Deviations compared

```python
std_cols = [col_name for col_name in df.columns if col_name.endswith("-mean_grouped_std_value") or col_name == "metric"]

#df_std = df[["metric", "gemma-mean_grouped_std_value", "llama-mean_grouped_std_value","mistral-mean_grouped_std_value"]].copy()
df_std = df[std_cols].copy()

def highlight_min(s, color=''):
    min = np.nanmin(
                [val for val in s.values if isinstance(val, Number)]
                 )
    return np.where(s==min, color, '')

df_std.style.apply(highlight_min, axis=1, color='background-color:lightgreen;')
```
